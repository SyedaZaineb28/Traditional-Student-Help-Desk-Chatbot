## Project Description

This project demonstrates the development of a traditional rule-based chatbot using Python. The chatbot answers frequently asked student questions by comparing user input with predefined keywords and regular-expression patterns. The project does not use a large language model or generative artificial intelligence.

## Project Objective

The objective of this project is to design, implement, and test a simple traditional chatbot that assists students with common academic and technical questions. The chatbot uses predefined rules, keyword matching, and regular expressions to select an appropriate response. It demonstrates the basic chatbot development lifecycle, including requirements analysis, conversation design, implementation, testing, evaluation, documentation, and deployment through GitHub.

## Chatbot Requirements

The chatbot should:

1. Greet the user.
2. Accept text input from the user.
3. Convert input to lowercase for easier comparison.
4. Recognize common greetings.
5. Answer questions about courses, assignments, professors, passwords, attendance, and technical support.
6. Respond politely when thanked.
7. End the conversation when the user types goodbye, quit, or exit.
8. Provide a fallback response when a question is not recognized.
9. Continue running until the user chooses to exit.
10. Use only traditional rule-based techniques.

```python
# Student Help Desk: Traditional Rule-Based Chatbot

def get_chatbot_response(user_message):
    """
    Examine the user's message and return a predefined response.
    This chatbot uses keyword matching and does not use an LLM.
    """

    # Normalize the input by removing extra spaces and converting it to lowercase.
    message = user_message.strip().lower()

    # Greeting rules
    if message in ["hello", "hi", "hey", "good morning", "good afternoon"]:
        return (
            "Hello! I am the Student Help Desk chatbot. "
            "How may I assist you today?"
        )

    # Course-information rules
    elif "course" in message or "class" in message:
        return (
            "You can find available courses and class information "
            "in the student portal or the official course catalog."
        )

    # Assignment rules
    elif "assignment" in message or "homework" in message:
        return (
            "Assignments are normally available in the course learning "
            "management system. Please open your course and check the "
            "Assignments section."
        )

    # Professor or instructor rules
    elif (
        "professor" in message
        or "instructor" in message
        or "teacher" in message
    ):
        return (
            "You can contact your professor using the email address "
            "listed in the course syllabus or learning management system."
        )

    # Password rules
    elif "password" in message or "login" in message:
        return (
            "Use the 'Forgot Password' option on the login page. "
            "Contact technical support if you cannot reset your password."
        )

    # Attendance rules
    elif "attendance" in message or "absent" in message:
        return (
            "Attendance requirements may differ by course. "
            "Please review the attendance policy in your syllabus."
        )

    # Technical-support rules
    elif (
        "technical support" in message
        or "computer problem" in message
        or "website not working" in message
    ):
        return (
            "Please restart the application, check your internet connection, "
            "and contact the university technical-support department if the "
            "problem continues."
        )

    # Appreciation rules
    elif "thank" in message:
        return "You are welcome! I am glad I could help."

    # Exit rules
    elif message in ["goodbye", "bye", "quit", "exit"]:
        return "Goodbye! Have a great day."

    # Fallback rule
    else:
        return (
            "I am sorry, but I do not understand that question. "
            "Please ask about courses, assignments, professors, passwords, "
            "attendance, or technical support."
        )


print("Student Help Desk Chatbot")
print("Type 'exit' when you want to end the conversation.")
print("-" * 50)

while True:
    user_input = input("You: ")

    chatbot_reply = get_chatbot_response(user_input)
    print("Chatbot:", chatbot_reply)

    if user_input.strip().lower() in ["goodbye", "bye", "quit", "exit"]:
        break
```

## Initial Test Observations

The first version of the chatbot executed successfully in Google Colab. The chatbot correctly recognized supported user inputs such as greetings and questions containing keywords including assignment, password, and attendance. It also terminated the conversation correctly when the user entered an exit command such as bye or exit.

Testing also demonstrated one of the primary limitations of a traditional rule-based chatbot. When the user entered a question outside the chatbot's predefined knowledge domain—for example, "What is the weather in Chicago?"—the chatbot was unable to provide a meaningful answer and instead displayed the fallback response. This behavior illustrates that rule-based chatbots rely entirely on predefined rules and keywords created by the developer, making their performance dependent on the completeness of the knowledge base.

```python
import re


def get_chatbot_response(user_message):
    """
    Return a predefined response using keywords and regular expressions.
    This chatbot does not use an LLM.
    """

    message = user_message.strip().lower()

    # Greeting patterns
    if re.search(r"\b(hello|hi|hey|good morning|good afternoon)\b", message):
        return (
            "Hello! I am the Student Help Desk chatbot. "
            "How may I assist you today?"
        )

    # Course and class information
    elif re.search(r"\b(course|courses|class|classes)\b", message):
        return (
            "You can find available courses and class information "
            "in the student portal or official course catalog."
        )

    # Assignment information
    elif re.search(r"\b(assignment|assignments|homework|project)\b", message):
        return (
            "Assignments are usually available in the learning management "
            "system. Open your course and check the Assignments section."
        )

    # Professor or instructor contact
    elif re.search(r"\b(professor|instructor|teacher|faculty)\b", message):
        return (
            "You can contact your professor using the email address "
            "listed in the course syllabus or learning management system."
        )

    # Password and login assistance
    elif re.search(r"\b(password|login|sign in|account access)\b", message):
        return (
            "Use the Forgot Password option on the login page. "
            "Contact technical support if the problem continues."
        )

    # Attendance questions
    elif re.search(r"\b(attendance|absent|absence|miss class)\b", message):
        return (
            "Attendance requirements differ by course. "
            "Please review the attendance policy in your syllabus."
        )

    # Grades
    elif re.search(r"\b(grade|grades|score|result)\b", message):
        return (
            "Grades are normally available in the course learning "
            "management system under the Grades section."
        )

    # Registration
    elif re.search(r"\b(register|registration|enroll|enrollment)\b", message):
        return (
            "Course registration is completed through the student portal. "
            "Contact your academic advisor if you need assistance."
        )

    # Library
    elif re.search(r"\b(library|book|books|database)\b", message):
        return (
            "You can access library resources, electronic books, and "
            "academic databases through the university library website."
        )

    # Technical support
    elif re.search(
        r"\b(technical support|computer problem|website not working|error)\b",
        message,
    ):
        return (
            "Check your internet connection and restart the application. "
            "Contact technical support if the issue continues."
        )

    # Thank-you pattern
    elif re.search(r"\b(thank you|thanks|thank)\b", message):
        return "You are welcome! I am glad I could help."

    # Exit patterns
    elif re.search(r"\b(goodbye|bye|quit|exit)\b", message):
        return "Goodbye! Have a great day."

    # Fallback response
    else:
        return (
            "I am sorry, but I do not understand that question. "
            "Please ask about courses, assignments, professors, passwords, "
            "attendance, grades, registration, the library, or technical support."
        )


print("Student Help Desk Chatbot")
print("Type 'exit' when you want to end the conversation.")
print("-" * 60)

while True:
    user_input = input("You: ")

    chatbot_reply = get_chatbot_response(user_input)
    print("Chatbot:", chatbot_reply)

    if re.search(r"\b(goodbye|bye|quit|exit)\b", user_input.strip().lower()):
        break
```

Additional responses were implemented for topics such as registration, grades, library resources, and technical support. Testing confirmed that the chatbot correctly recognized a wider range of user inputs while continuing to provide a fallback response for unsupported questions. These improvements increased the chatbot's usability while maintaining the traditional rule-based design.
